```
import pandas as pd
import matplotlib.pyplot as plt

# Datos simulados para esta ocasión
data = {
    "cliente": ["Ana", "Luis", "Pedro", "Ana", "Luis", "Pedro"],
    "fecha_numero": [15032024, 5072024, 12042024, 23052024, 11062024, 25062024],
    "valor_compra": [120000, 85000, 64000, 150000, 90000, 72000]
}

df = pd.DataFrame(data)

print("Datos originales:")
print(df)

# -------------------------------
# Función para convertir número a fecha
# -------------------------------

def convertir_fecha(numero):
    texto = str(numero)

    # Si la fecha tiene 7 dígitos le agregamos un cero al día
    if len(texto) == 7:
        texto = "0" + texto

    return pd.to_datetime(texto, format="%d%m%Y")

df["fecha"] = df["fecha_numero"].apply(convertir_fecha)

print("\nDatos con fecha convertida:")
print(df)

# -------------------------------
# Análisis de datos
# -------------------------------

ventas_por_cliente = df.groupby("cliente")["valor_compra"].sum()

print("\nVentas totales por cliente:")
print(ventas_por_cliente)

# -------------------------------
# Gráfico de ventas
# -------------------------------

ventas_por_cliente.plot(kind="bar")

plt.title("Ventas totales por cliente")
plt.xlabel("Cliente")
plt.ylabel("Valor de ventas")
plt.show()
```
